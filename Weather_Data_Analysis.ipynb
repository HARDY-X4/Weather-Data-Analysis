```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'],
    'Temperature': [30, 32, 31, 29, 28],
    'Humidity': [70, 65, 75, 80, 85],
    'Condition': ['Sunny', 'Sunny', 'Cloudy', 'Rainy', 'Rainy']
}
df = pd.DataFrame(data)


df['Date'] = pd.to_datetime(df['Date'])


plt.figure(figsize=(8, 5))
plt.plot(df['Date'], df['Temperature'], marker='o', label='Temperature')
plt.title('Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid()
plt.show()


plt.figure(figsize=(6, 4))
sns.barplot(x='Condition', y='Humidity', data=df)
plt.title('Average Humidity by Weather Condition')
plt.xlabel('Condition')
plt.ylabel('Humidity (%)')
plt.show()


print("Summary Statistics:")
print(df.describe())


max_temp_day = df.loc[df['Temperature'].idxmax()]
print(f"The hottest day was {max_temp_day['Date'].date()} with a temperature of {max_temp_day['Temperature']}°C.")
```
